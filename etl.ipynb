```python
import pandas as pd

dataframe = pd.read_csv("data/mental_health.csv")

#visualizações da tabela
#dataframe.head(3)
#dataframe.info()
#dataframe.describe()
dataframe.isnull().sum()


'''for col in dataframe.columns: #codigo utilizado para verificar os valores únicos de cada coluna
    print(f"\n{col}:")
    print(dataframe[col].unique()[:10])'''
```

```python
dataframe = dataframe.drop_duplicates()#removendo linhas duplicadas
dataframe = dataframe.drop_duplicates(subset='Person_ID')#removendo linhas duplicadas por ID

dataframe = dataframe[(dataframe['Age'] > 0) & (dataframe['Age'] < 100)]#removendo idades inválidas

dataframe = dataframe[(dataframe['Daily_Screen_Time'] >= 0) & (dataframe['Daily_Screen_Time'] <= 24)]#removendo tempos de tela inválidos
dataframe = dataframe[(dataframe['Sleep_Hours'] >= 0) & (dataframe['Sleep_Hours'] <= 24)]#removendo horas de sono inválidas

cols_binarias = ['Depression', 'Anxiety', 'Burnout', 'Smoking', 'Alcohol']#verificando se as colunas binárias possuem apenas 0 e 1

for col in cols_binarias:# removendo valores inválidos das colunas binárias
    dataframe = dataframe[dataframe[col].isin([0,1])]

dataframe['Gender'] = dataframe['Gender'].str.strip().str.capitalize()#padronizando os valores de gênero
dataframe['Occupation'] = dataframe['Occupation'].str.strip().str.capitalize()#ocupação


# Faixa etária
dataframe['Age_Group'] = pd.cut(# criando faixas etárias
    dataframe['Age'],
    bins=[0,18,25,35,50,100],
    labels=['Teen','Young Adult','Adult','Mid Age','Senior']
)

# Tela
dataframe['Screen_Group'] = pd.cut(
    dataframe['Daily_Screen_Time'],
    bins=[0,4,6,8,12],
    labels=['Low','Moderate','High','Very High']
)

# Sono
dataframe['Sleep_Group'] = pd.cut(
    dataframe['Sleep_Hours'],
    bins=[0,5,6,7,9],
    labels=['Very Low','Low','Ideal','High']
)

dataframe.to_csv("mental_health_clean.csv", index=False)#salvando o dataframe limpo em um novo arquivo CSV
```

```python

for col in ['Gender', 'Occupation', 'Physical_Activity']: # codigo utilizado para verificar a contagem de cada valor entender distribuição
    print(f"\n{col}:")
    print(dataframe[col].value_counts())

dataframe[['Depression', 'Anxiety', 'Burnout']].value_counts() #conta as combinações ex 1 ou 0 para cada uma das colunas

dataframe[['Depression', 'Anxiety', 'Burnout']].describe() # estatísticas descritivas dos indicadores

dataframe.groupby('Gender')[['Depression', 'Anxiety', 'Burnout']].mean() # média de cada indicador por gênero


dataframe.groupby(pd.cut(dataframe['Daily_Screen_Time'], bins=5))[ #agrupamento por faixas de tempo de tela
    ['Depression', 'Anxiety', 'Burnout']
].mean()

dataframe.groupby(pd.cut(dataframe['Sleep_Hours'], bins=5))[ #agrupamento por faixas de horas de sono
    ['Depression', 'Anxiety', 'Burnout']
].mean()

dataframe.groupby('Occupation')[ #agrupamento por ocupação
    ['Depression', 'Anxiety', 'Burnout']
].mean()
```
