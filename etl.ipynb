```python

dataframe = dataframe.drop_duplicates()
dataframe = dataframe.drop_duplicates(subset='Person_ID')

dataframe = dataframe[(dataframe['Age'] > 0) & (dataframe['Age'] < 100)]

dataframe = dataframe[(dataframe['Daily_Screen_Time'] >= 0) & (dataframe['Daily_Screen_Time'] <= 24)]
dataframe = dataframe[(dataframe['Sleep_Hours'] >= 0) & (dataframe['Sleep_Hours'] <= 24)]

cols_binarias = ['Depression', 'Anxiety', 'Burnout', 'Smoking', 'Alcohol']

for col in cols_binarias:
    dataframe = dataframe[dataframe[col].isin([0,1])]

dataframe['Gender'] = dataframe['Gender'].str.strip().str.capitalize()
dataframe['Occupation'] = dataframe['Occupation'].str.strip().str.capitalize()
dataframe['Physical_Activity'] = dataframe['Physical_Activity'].str.strip().str.capitalize()


# Faixa etária
dataframe['Age_Group'] = pd.cut(
    dataframe['Age'],
    bins=[0,18,25,35,50,100],
    labels=['Teen','Young Adult','Adult','Mid Age','Senior']
)

# Tela
dataframe['Screen_Group'] = pd.cut(
    dataframe['Daily_Screen_Time'],
    bins=[0,4,6,8,12],
    labels=['Low','Moderate','High','Very High']
)

# Sono
dataframe['Sleep_Group'] = pd.cut(
    dataframe['Sleep_Hours'],
    bins=[0,5,6,7,9],
    labels=['Very Low','Low','Ideal','High']
)

dataframe.to_csv("mental_health_clean.csv", index=False)
```
